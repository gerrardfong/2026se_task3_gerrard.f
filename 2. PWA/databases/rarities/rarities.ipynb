```python
text = """ (Useless, Flawed, Legendary),
    (Subhuman, Flawed, Rare),
    (Chudlite, Human, Rare),
    (Average, Human, Common),
    (Athletic, Human, Common),
    (Building, Superhuman, Uncommon),
    (City, Urban, Rare),
    (Mountain, Urban, Rare),
    (Island, Nuclear, Epic),
    (Country, Nuclear, Epic),
    (Continental, Tectonic, Epic),
    (Planet, Planetary, Legendary),
    (Star, Planetary, Legendary),
    (Solar System, Stellar, Legendary),
    (Galaxy, Stellar, Mythic),
    (Universal, Universal, Mythic),
    (Multiversal, Multiversal, Mythic),
    (Outerversal, Outversal, Mythic),
    (Absolute, Boundless, Ultra) """

result_lines = []
for line in text.strip().split('\n'):
    line = line.strip()
    has_comma = line.endswith(',')
    if has_comma:
        line = line[:-1]
    elements = line.strip('()').split(',')
    quoted_elements = []
    for element in elements:
        quoted_elements.append(f"'{element.strip()}'")
    result_line = '(' + ', '.join(quoted_elements) + ')' + (',' if has_comma else '')
    result_lines.append(result_line)

print('\n    '.join(result_lines))
```
