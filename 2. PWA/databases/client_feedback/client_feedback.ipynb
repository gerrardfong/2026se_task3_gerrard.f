```python
%%sql

CREATE TABLE IF NOT EXISTS client_feedback(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    date DATE NOT NULL,
    email TEXT NOT NULL,
    types TEXT NOT NULL,
    feedback TEXT
);



INSERT INTO client_feedback (date, email, types ,feedback) VALUES(
    "2026/04/29","van.pham167@education.nsw.gov.au","Bug","The game doesn't run"
);

SELECT * FROM client_feedback;
```

```python
# Function to separate CSV data into tuples for SQL

text = "2026/04/29 10:24:04 PM GMT+10	van.pham167@education.nsw.gov.au	Bug	The game doesn't run"


tuples = []

for line in text.strip().splitlines():
    line = line.replace("\t", ",")
    date, email, types, feedback = line.split(',')
    date=date[:10]
    tuples.append((date, email, types, feedback))

print(tuples)
```
