```python
%%sql

CREATE TABLE IF NOT EXISTS bosses (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL UNIQUE,
    pfp TEXT,
    tier TEXT NOT NULL,
    xp INTEGER NOT NULL,
);

CREATE TABLE IF NOT EXISTS boss_attributes (
    boss_id INTEGER NOT NULL REFERENCES bosses(id),
    attribute TEXT NOT NULL REFEREBCES attributes(attribute),
    level INTEGER NOT NULL REFERENCES rarities(level),
    PRIMARY KEY (boss_id, attribute)
);
```
