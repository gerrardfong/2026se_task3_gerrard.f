```python
%%sql

CREATE TABLE IF NOT EXISTS rarities (
    level TEXT PRIMARY KEY,
    rarity TEXT NOT NULL
);

SELECT * FROM rarities;
```

```python
%%sql

INSERT OR IGNORE INTO rarities (level, rarity) VALUES 
    ("Average", "Common"),
    ("Athletic", "Common"),
    ("Building", "Uncommon"),
    ("Subhuman", "Rare"),
    ("Chudlite", "Rare"),
    ("City", "Rare"),
    ("Mountain", "Rare"),
    ("Island", "Epic"),
    ("Country", "Epic"),
    ("Continental", "Epic"),
    ("Planet", "Legendary"),
    ("Star", "Legendary"),
    ("Solar System", "Legendary"),
    ("Galaxy", "Legendary"),
    ("Useless", "Legendary"),
    ("Universal", "Mythic"),
    ("Multiversal", "Mythic"),
    ("Outerversal", "Ultra"),
    ("Absolute", "Ultra");

SELECT * FROM rarities;
```

```python
import sqlite3

RARITY_WEIGHTS = {
    'Common':    40,
    'Uncommon':  25,
    'Rare':      15,
    'Epic':       9,
    'Legendary':  5,
    'Mythic':     4,
    'Ultra':      2,
}

# Normalise weights to percentages
total = sum(RARITY_WEIGHTS.values())
rarity_pct = {rarity: round(weight / total * 100) for rarity, weight in RARITY_WEIGHTS.items()}

# Attach percentages to each level from the DB
conn = sqlite3.connect('characters.db')
rows = conn.execute("SELECT level, rarity FROM rarities").fetchall()

for level, rarity in rows:
    pct = rarity_pct.get(rarity, 0)
    print(f"{level} [{rarity}] — {pct}%")

conn.close()
```
