## Design

### Business Problemn

I'd like to create a game for my client that simulates a random generation system that is used to create a character. The game will be RPG and include a looting and experience system. I plan to do Canary Release using beta-testers who all have different purposes to make my game as versatile as possible. Realistically in the end, the game is technically made my client and I therefore it doesn't necessarily have to go public.

### Success Metrics

Some requirements or measures I plan to involve:

- Game requirements:
    - Primarily start off with **PVE** and potentially develop **PvP** after deploying the game, although the focus for this assignment will be on PvE
    - Hopefully some sort of closed loop machine-learning system that allows for the enemies in my game to adapt based off of user inputs
    - Implement a training feature as that means **characters** will permanently be stuck at their created level (Discuss with client)
    - Combat:
        - Main combat system will be comprised of comparing character *attributes* and then determining who wins by who has better attributes
        - Implement a **turn-based** system where character attributes are linked to values which then can be used to determine health, attack power and other characteristics
        - *Bosses* in the game be based off of existing characters

- User requirements:
    - Players should be able to fight and level up their character, the game won't be designed to grind and will most likely have an end phase or be repetition of fighting
    - Players should be able to create (name, generate attributes), delete, sort, train their characters
    - Players should be able to fight other players or **NPCs**

- System requirements:
    - Game functionality:
        - **RNG** system that will most likely use pre-made functions and will require imports to randomise possibilities
        - Use hard-coded **rules** that will compare character vs enemy attributes to decide who wins
        - Use a feedback loop that will train the Bosses on user patterns (attack after they heal)
    - Databases: 
        - Use SQL databases or multidimensional dictionaries to create categories for RNG generation
        - SQL database that will store user information and build information for characters
        - Use joins for different attributes and possible outcomes 
    - UI:
        - The base UI will consist of two separate sections: Characters, Fighting
        - Use **spinning wheels** when determining attributes for character creation
        - The game will have a UI that will most likely be AI'd with some personal tweaks to create the most user-friendly layout, utilising CSS and js for optimisation
    - Client Feedback:
        - Link Google Forum responses into an admin tab where data can be collected and reviewed
    
    
### Glossary: Domain Knowledge

| Word | Definition |
| - | - |
| PVE | Player Versus Entity, When a player goes against an enemy that is in-built into the system |
| PVP | Player Versus Player, When players are placed against one another |
| characters | An entity within the game that contains specific attributes creating the identity of the character |
| attributes | Mandatory characteristics that create the foundation of the character and allow it to compete against other characters |
| turn-based | A unique style of combat where each player takes turns to choose an action until one player either loses or leaves |
