## Design

### Business Problemn

I'd like to create a game for my client that simulates a random generation system that is used to create a character. The game will be RPG and include a looting and experience system. I plan to do Canary Release using beta-testers who all have different purposes to make my game as versatile as possible. Realistically in the end, the game is technically made my client and I therefore it doesn't necessarily have to go public.

### Success Metrics

Some requirements or measures I plan to involve:

- Game requirements:
    - Primarily start off with PVE and potentially develop PvP after deploying the game, although the focus for this assignment will be on PvE
    - Hopefully some sort of closed loop machine-learning system that allows for the enemies in my game to adapt based off of user inputs

- User requirements:
    - Players should be able to fight and level up their character, the game won't be designed to grind and will most likely have an end phase or be repetition of fighting

- System requirements:
    - SQL database that will store user information and build information for characters
    - RNG system that will most likely use pre-made functions and will require imports to randomise possibilities
    - Use SQL databases or multidimensional dictionaries to create categories for RNG generation
    - The game will have a UI that will most likely be AI'd with some personal tweaks to create the most user-friendly layout, utilising CSS and js for optimisation
